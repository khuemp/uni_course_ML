$gaussian=\frac{1}{\sqrt{2 \pi \sigma^2}} e^-\frac{(x-\mu)^2}{2\sigma^2}$

```python
def gaussian(x, mu, var):
    first = 1 / np.sqrt(2 * np.pi * var)
    second = np.exp(-0.5 * ((x-mu)**2) / var)
    return first * second
```

```python
p = np.linspace(-4, 4, num=1000)
plt.plot(p, gaussian(p, mu=1, var=1), c="b", label="$\mu=1,\sigma^2=1$")
plt.plot(p, gaussian(p, mu=-1, var=1), c="g", label="$\mu=-1,\sigma^2=1$")
plt.plot([0]*1000, gaussian(p, -1, 1), c="r")
plt.legend()
plt.show()
```

# Polynomial curve fitting

```python
np.random.seed(1234) # for reproducibility

def func(x):
    return np.sin(2 * np.pi * x)

def create_data(func, sample_size, std):
    x = np.linspace(0, 1, sample_size)
    t = func(x)
    t = t + np.random.normal(scale=std, size=x.shape) # add noise
    return x, t


x_train, y_train = create_data(func, sample_size=10, std=0.25)
x_test, y_test = create_data(func, sample_size=100, std=0.00)

plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, c="g", label="$\sin(2 \pi x)$")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

plt.figure(figsize=(15, 10))

for i, degree in enumerate([0, 1, 3, 9]):
    plt.subplot(2, 2, i+1)
    
    features = PolynomialFeatures(degree).fit(x_train[:, None])
    X_train = features.transform(x_train[:, None])
    X_test = features.transform(x_test[:, None])
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
    plt.plot(x_test, y_test, c="g", label="$\sin(2\pi x)$")
    plt.plot(x_test, y_pred, c="r", label="predicted")
    plt.ylim(-1.5, 1.5)
    plt.legend(loc=0)
    plt.title("degree={}".format(degree))
    
plt.show()
```

# Root Mean Square Error (RMSE)

```python
def rmse(a, b):
    return np.sqrt(np.mean(np.square(a - b)))

training_errors = []
test_errors = []

for i in range(10):
    feature = PolynomialFeatures(i).fit(x_train[:, None])
    X_train = feature.transform(x_train[:, None])
    X_test = feature.transform(x_test[:, None])
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    y = model.predict(X_test)
    training_errors.append(rmse(model.predict(X_train), y_train))
    test_errors.append(rmse(y, y_test))
    
plt.figure()
plt.plot(training_errors, "o-", mfc="none", mec="b", ms=10, c="b", label="Training")
plt.plot(test_errors, "o-", mfc="none", mec="r", ms=10, c="r", label="Testing")
plt.legend()
plt.xlabel("degree")
plt.ylabel("RMSE")
plt.xlim(-1, 10)
plt.show()
```

```python
from sklearn.linear_model import Ridge as RidgeRegression

feature = PolynomialFeatures(9).fit(x_train[:, None])
X_train = feature.transform(x_train[:, None])
X_test = feature.transform(x_test[:, None])

model = RidgeRegression(alpha=1e-4) # alpha is the regularization term
model.fit(X_train, y_train)
y = model.predict(X_test)

plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, c="g", label="$\sin(2\pi x)$")
plt.plot(x_test, y, c="r", label="fitting")
plt.ylim(-1.5, 1.5)
plt.legend()
plt.show()
```

# Central Limit Theorem

## What does it say?

The theorem states that, under various mild conditions, the distribution of the average of a large number of independent random variables converges to the normal distribution, regardless

Given N independently and identically distributed (iid) random variablex $X_1$ to $X_n$, and let their common distribution have finite expection $\mu$ and finite variance $\sigma^2 > 0$.

We form a new random variable $Z_n$ by averaging: $Z_n = \frac{1}{n} \sum{_{i=1}^n X_i}$.

According to central limit theorem:

1. $Z_n$ has expectation $\mu$ $\rightarrow$ $\mathbb{E}\big[\frac{1}{n}\sum{_{i=1}^n X_i}\big] = \frac{1}{n}\big[\sum{_{i=1}^n \mathbb{E}[X_i]}\big] = \frac{1}{n}*n*\mu = \mu$

2. $Z_n$ has variance $\sigma^2/n$ $\rightarrow$ $Var\big(\frac{1}{n}\sum{_{i=1}^n X_i} = \frac{1}{n^2}\sum{_{i=1}^n  Var(X_i)}\big) = \frac{1}{n^2}*n*\sigma^2 = \frac{1}{n}\sigma^2$

3. Distribution of $Z_n$ converges to a normal distribution as $n\rightarrow \infty$ 

And thus, if we form $Z_n^* = \frac{Z_n - \mathbb{E}[Z_n]}{\sqrt{Var(Z_n)}} = \frac{Z_n - \mu}{\sigma/\sqrt{n}} = \sqrt{n} * \frac{Z_n - \mu}{\sigma}$, this will, as $n \rightarrow \infty$, converge to N(0, 1)

## Let's start by plotting the underlying ("population") distribution
For example, we've chosen a mixture of two gaussians, but any distribution can be assumed, as long as its mean and variance are finite

```python
import scipy.stats as stats
import random

def sample_from_gaussian_mixture(num_samples,  centers, std, weight=0.5):
    """
    Generates data from a K=len(centers) GMM where the i-th cluster is a N(center[i], std)
    and num_samples samples is generated from it.
    
    Returns a numpy array of shape (len(centers) * num_samples)
    """
    
    mask = (np.random.random(size=num_samples) < weight)
    
    samples1 = np.random.normal(centers[0], std, num_samples)
    samples2 = np.random.normal(centers[1], std, num_samples)
    
    return (mask*samples1 + (1-mask)*samples2).reshape(-1)
```

```python
linspace = np.linspace(-7.5, 7.5, 1000)
plt.figure(figsize=(10, 5))

centers = [-2, 2]
std=1
weight=0.5
num_samples=5000

X = sample_from_gaussian_mixture(num_samples, centers=centers, std=std, weight=weight)
mixture_pdf = weight*stats.norm.pdf(linspace, centers[0], std) + (1-weight)*stats.norm.pdf(linspace, centers[1], std)

plt.plot(linspace, mixture_pdf, color="red")
plt.hist(X, density=True, bins=round(num_samples/100), color="blue", alpha=0.5)
plt.show()
```

## And now, lets plot our sampling distribution:

```python
N=20

def sample_from_sampling_distribution(population_samples, centers, std, weight): # evaluates Z_N^*
    # mean of Z_N
    mu = (weight*centers[0] + (1-weight)*centers[1])
    
    # variance of our population distribution, see https://en.wikipedia.org/wiki/Mixture_distribution
    var = ((weight*(std**2+(centers[0])**2)) + ((1-weight)*(std**2+(centers[1])**2))) - mu  
    
    std_zn = np.sqrt(var) / np.sqrt(N) # std-dev of Z_N
    
    Z_n = np.mean(population_samples)
    Z_n_star = (Z_n - mu) / std_zn
    return Z_n_star
```

```python
num_sampling_distribution_samples = 5000
sampling_distribution_samples = []

for _ in range(num_sampling_distribution_samples):
    population_samples = sample_from_gaussian_mixture(N, 
                                                      centers=centers, 
                                                      std=std, 
                                                      weight=weight)
    sampling_distribution_sample = sample_from_sampling_distribution(population_samples, 
                                                                     centers=centers, 
                                                                     std=std, 
                                                                     weight=weight)
    sampling_distribution_samples.append(sampling_distribution_sample)
```

```python
plt.hist(sampling_distribution_samples, density=True, bins=50)
plt.plot(linspace, stats.norm.pdf(linspace, 0, 1), color="red")
plt.show()
```
