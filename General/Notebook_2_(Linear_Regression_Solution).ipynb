## Generating some data

First, we are going to create the functions to generate linear and sinusoidal data in the interval $x \in [0,1]$, adding gaussian noise.

**Linear data:** $y = wx + b + \epsilon$ with $\epsilon \sim \mathcal{N}(0, \sigma)$

**Sinosoidal data:** $y = sin(2\pi x) + \epsilon$ with $\epsilon \sim \mathcal{N}(0, \sigma)$

$\sim$ means "distributed according to"

```python
# Generate data in the interval [0, 1]
def make_linear_data(num_points, w, b, noise_std):
    x = np.sort(np.random.uniform(low=0, high=1, size=num_points)) # result is of shape (num_points,) and sorted in ascending order
    epsilon = np.random.normal(loc=0, scale=noise_std, size=num_points) 
    y = (w * x) + b + epsilon
    return x, y

def make_sin_data(num_points, noise_std):
    x = np.sort(np.random.uniform(low=0, high=1, size=num_points))
    epsilon = np.random.normal(loc=0, scale=noise_std, size=num_points)
    y = np.sin(2 * np.pi * x) + epsilon
    return x, y
    
w_true = 2
b_true = 1
# np.random.seed(42) # for reproducibility

x_linear, y_linear = make_linear_data(100, w=w_true, b=b_true, noise_std=0.2)
x_sinus, y_sinus = make_sin_data(100, noise_std=1)
    
```

To visualize the datasets these functions produce, we can plot them:

## Least Squares Regression (LSR)

Given a dataset with $n$ i.i.d training samples $D = (X_i, Y_i)_{i=1}^n$ and a function space $\mathcal{F}$ we define the least squares regression solution as 
$$f_n^* = \underset{f \in \mathcal{F}}{\arg \min} \frac{1}{n}\sum_{i}^{n}(y_i - f(x_i))^2$$

We also know that the optimal solution to LSR, i.e, the Bayes optimal function, is given by 
$$f^*(\mathbf{x}) = \mathbb{E}[Y|X=\mathbf{x}]$$

Here, we are looking at linear LSR, which assumes that the function class consists of all linear functions.
$$\mathcal{F} = \bigg\{f: \mathbb{R}^M \rightarrow \mathbb{R}, f(\mathbf{x}) = \sum_{i=1}^{M} w_i \phi _i(\mathbf{x}) \mid \mathbf{w} \in \mathbb{R}^M \bigg\}$$

where $\phi_i$ are called basis functions $\phi_i: \mathbb{R}^d \rightarrow \mathbb{R}$ and represent transformations of the input data. Some examples of basis functions are 
$$\phi(x) = \big(1, x, x^2, ...\big), \phi(x)=\big(sin(x), cos(x), ...\big)$$

Note that, although the basis functions produce vectors with entries that are **non-linear** with respect to their input, the function class still only contains **linear** functions i.e linear in the parameters $w$.

### LSR solution
As seen in the lecture, we have an explicit solution:

Let $X \in \mathcal{R}^{nxM}$. Here, we assume that the first column of this matrix is full of ones, this will make the calculation easier. 

Following the lecture slide, we now simply compute the optimal weight: $w^* = (X^TX)^{-1}X^Ty$. Note that we can only compute the inverse of $X^TX$, if it is of full column rank. Otherwise, we must resort to the pseudo-inverse. This problem can be circumvented by using the numpy function `pinv`, which equals the normal inverse if the input is invertible and returns the pseudoinverse otherwise.

```python
def LSR_optimal_w_and_b(X, y):
    X = np.vstack([np.ones(len(X)), X]).T # add ones to first column
    XTX = X.T @ X
    XTX_inverse = np.linalg.pinv(XTX)
    b, w = XTX_inverse @ (X.T @ y) # extract intercept and w-value from the solution
    
    return b, w
```

Now we can apply our function to our datasets and plot the resulting solutions

## Noise-Bias-Variance Decomposition

The risk (i.e, the expected squared loss) $R(f_n)$ of the estimator $f_n$ is given by:

$$R(f_n) = \mathbb{E}\big[(Y - f_n(X))^2\big] = \mathbb{E}\big[(Y - \mathbb{E}[Y|X])^2\big] + \mathbb{E}\big[ (\mathbb{E}[Y|X] - f_n(X))^2 \big]$$

In this first case, we left the estimator $f_n$ fixed. But, since $f_n$ is dependent on the sampled dataset $X$, we can also look at the "expected risk" by taking into considerations the probabilities with which we sample every possible dataset of size n:

\begin{align}
\mathbb{E_D} [R(f_n)] = \mathbb{E} \big[ (Y - f^{*}(X))^2 \big]  + \mathbb{E}_D \big[ \mathbb{E}_X \big[ (f^*(X) - f_n(X))^2 \big] \big]\\
= \mathbb{E} \big[ (Y - f^{*}(X))^2 \big]  + \operatorname{Var}(f_n(X)) + (\operatorname{Bias}(f_n(X)))^2 \\
= \sigma^2 + \operatorname{Var}(f_n(X)) + (\operatorname{Bias}(f_n(X)))^2
\end{align}

Here, we wrote $f^*(X)$ instead of $\mathbb{E}[Y|X]$ for easier readability. For a derivation of these equation, refer to the lecture slides. The $\sigma$ corresponds to the standard deviation we used to generate the noise.

**KEY POINT 1.** The first term is called "noise term" or "irreducible error". It is constant, i.e. it doesn't depend on the sampled dataset nor on the estimator.

**KEY POINT 2.** The trade-off between (squared) bias and variance corresponds to the trade-off between overfitting and underfitting.

### Relation to regularization
This is also related to regularization: when using regularization in least squares regression (e.g as we do in Lasso or Ridge Regression), we will introduce some bias (whereas the ordinary least squares regression provides unbiased estimate) but in turn we can greatly reduce variance and thus produce superior performance in general.

## Linear data with varying amount of noise

Intuitively, it should be clear that our estimator has larger variance if either the variance of the noise increases or the number of samples in the dataset decrease.

We can test this intuition by looking at the best value for $w$ and $b$, as well as the MSE, for varying levels of noise and dataset sizes:

```python
# Mean Squared Error (MSE) a.k.a Risk of squared loss
def MSE(ground_truth, prediction):
    return np.mean(np.square(ground_truth - prediction))
```

```python
# one could also show that even for large variance, the correct coeff are found as n -> infinty
num_samples = 6
noise_std = 2
x_train, y_train = make_linear_data(num_samples, w=2, b=1, noise_std=noise_std)
reg = LinearRegression().fit(x_train.reshape(-1, 1), y_train)
x_test, y_test = make_linear_data(num_samples*10000, w=2, b=1, noise_std=noise_std)

prediction = reg.predict(x_test.reshape(-1, 1))
loss = MSE(y_test, prediction)
print(f"w_star: {reg.coef_[0]:.3f}")
print(f"b_star: {reg.intercept_:.3f}")
print(f"MSE: {loss:.3f} | irreducible error: {noise_std**2:.3f}")
```

### Visualizing the distribution of $w*$ and $b*$
To get a better look at how much the $w*$ and $b*$ vary with different datasets, we can run a more thorough experiment:

Questions we might ask is:
1. Is out estimator unbiased?

   -> Do our samples on expectation match the true w and b that were used for generating the data?
2. How does the estimator change with n (dataset size)

   -> We can plot a histogram to find out how it is distributed

```python
w_true, b_true, noise_std = 2, 1, 1

N = 10 # Number of datapoints per dataset

W_stars, B_stars = [], []

for m in range(5000): # Changing this value does not change the distribution.
    x_data, y_data = make_linear_data(N, w_true, b_true, noise_std)
    reg = LinearRegression().fit(x_data.reshape(-1, 1), y_data)
    W_stars.append(reg.coef_[0]) # Extract w_star from the result
    B_stars.append(reg.intercept_) # Extract b_star from the result
    
print(f"E[w*] is approximately {np.mean(W_stars):.3f}, the true value is {w_true}. Var[w*] is approximately {np.var(W_stars, ddof=1):.3f}")
print(f"E[b*] is approximately {np.mean(B_stars):.3f}, the true value is {b_true}. Var[b*] is approximately {np.var(B_stars, ddof=1):.3f}")
```

```python
# Visualize the result
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# TODO: Plot the histogram for W_stars
ax[0].hist(W_stars, bins=25, density=False)
ax[0].axvline(w_true, c="r", linestyle="dashed", linewidth=2)
ax[0].set_xlabel("w*");

# TODO: Plot the histogram for B_stars
ax[1].hist(B_stars, bins=25, density=False)
ax[1].axvline(b_true, c="r", linestyle="dashed", linewidth=2)
ax[1].set_xlabel("b*");

# Fix the axis such that the plots are always comparable with new data
ax[0].set_xlim([0, 4]);
ax[1].set_xlim([0, 2]);
```

## Introducing basis functions
Basis functions can be used to apply linear regression methods, even though the relationship between the input features $X$ and the output target $Y$ is *not* linear. We do this by transforming our input feature into a new space (of generally higher dimension) in which the relationship between the input and output is now linear.

**Example:** For example, we can use polynomials as the basis function: $\phi(x) = [1, x, x^2, ..., x^p]$. For a 1-D regression problem, $X$ would be $\in \mathcal{R}^{nx1}$. If we apply the above polynomial basis function (for some finite $p$), the new, "transformed" dataset will be of shape $X \in \mathcal{R}^{n\mathsf{x}(p+1)}$. The original one-dimensional datapoint (e.g 2) would be transformed to the $(p+1)$-dimensional datapoint $(1, 2, 4, ..., 2^p)$.

**Questions we might ask:**
1. How does the training/test error vary with $p$?

   -> We can plot the MSE's (predictions vs. ground truth, for both train and test data) on the y-Axis, and $p$ on the x-Axis.

2. How can we relate what we find to the concept of under-/overfitting and the bias-variance trade-off?
3. How can we pick a suitable value for $p$, without relaying on the test data?

Side note: Recall that addressed all these questions in programming exercise notebook 1. We will continue to build upon what we learnt.

To find out more, we first write a function that gives us, for all degrees up to some limit, the train and test errors:

```python
# Parameter generator: a function that takes in a size (n) and returns a dataset
# Parameter regressor: either LinearRegression or Ridge or Lasso (all are classes for skleanrn)

def evaluate_models(max_p, train_data, test_data, regressor):
    x_train, y_train = train_data
    x_test, y_test = test_data
    
    training_errors, test_errors, models = [], [], []
    for i in range(max_p+1):
        # The [:, None] part converts from shape (n,) to shape (n, 1)
        features = PolynomialFeatures(i+1).fit(x_train[:, None])
        X_train = features.transform(x_train[:, None])
        X_test = features.transform(x_test[:, None])
        
        model = regressor().fit(X_train, y_train)
        models.append(model)
        
        training_errors.append(MSE(model.predict(X_train), y_train))
        test_errors.append(MSE(model.predict(X_test), y_test))
    return training_errors, test_errors, models
```

Now we can plot the MSE values (y-Axis) with the values for p(x-Axis):

```python
train_size, test_size = 50, 100
w_true, b_true, std = 2, 1, 0.1

# np.random.seed(42) # fix randomness

# generator is now a function taking only one argument (size) and passing it to make_linear_data
generator = lambda size: make_linear_data(size, w_true, b_true, std)
regressor = LinearRegression

train_data = generator(train_size)
test_data = generator(test_size)

max_p = 10
training_errors, test_errors, models = evaluate_models(max_p, train_data, test_data, regressor)

# TODO: Plot the result
plt.figure(figsize=(10, 5))
plt.plot(training_errors, "bo-", label="Training")
plt.plot(test_errors, "ro-", label="Testing")
plt.legend()
plt.xlabel("degree")
plt.ylabel("MSE")
plt.xlim(-1, max_p+1);
```

For the higher values of p, we observe overfitting!

## Model selection using the sinusoidal data
We can perform similar analysis for the sinusoidal data

```python
def plot_error_curves(train_errors, test_erros, ax=None, max_p=10):
    if ax is None:
        ax = plt.gca()
    ax.plot(train_errors, "bo-", label="Training")
    ax.plot(test_errors, "ro-", label="Testing")
    ax.legend()
    ax.set_xlabel("degree")
    ax.set_ylabel("MSE")
    ax.set_xlim(-1, max_p+1)
```

```python
std = 0.15
train_size, test_size = 20, 50
# np.random.seed(42) # fix randomness

generator = lambda size: make_sin_data(size, std)
regressor = LinearRegression
train_data = generator(train_size)
test_data = generator(test_size)

max_p = 11
training_errors, test_errors, models = evaluate_models(max_p, train_data, test_data, regressor)
plot_error_curves(training_errors, test_errors, max_p=max_p)
```

From these plots, we see that we can indeed prevent overfitting when using regularization.
